# Gradient Bandit

```python
%load_ext autoreload
%autoreload 2
from nArmedBandit import run_Gradient_bandit, init_baseline
import numpy as np
import matplotlib.pyplot as plt

num_steps = 1000
n = 10
step_size = 0.1
num_runs = 200
avg_rewards = np.zeros(num_steps)
avg_optimal = np.zeros(num_steps)
for i in range(num_runs):
    rewards, optimal = run_Gradient_bandit(init_baseline, n=n, alpha=step_size,num_steps=num_steps)
    avg_rewards += (rewards - avg_rewards)/(i+1)
    avg_optimal += (optimal - avg_optimal)/(i+1)

runs = np.arange(num_steps)
plt.plot(runs[::10],avg_optimal[::10])
plt.show()
```
